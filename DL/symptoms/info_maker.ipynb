```python
import json

with open('drive/MyDrive/disease/release_evidences.json', 'r') as fcc_file:
    fcc_data = json.load(fcc_file)
    print(fcc_data)
```

```python
list_of_items = list(fcc_data.keys())
```

```python
from natsort import natsorted
list_of_items = natsorted(list_of_items)
```

```python
dt = {}
dt_str_end = {}
dt_list = []
count = 0
for i in list_of_items:
  dt_str_end[i] = count
  temp_ver = {}
  temp_index = {}
  if len(fcc_data[i]["possible-values"]) != 0:
    if len(fcc_data[i]["value_meaning"]) == 0:
      temp_ver = fcc_data[i]["possible-values"]
      temp_index = None
    else:
      for k, n in zip(fcc_data[i]["possible-values"], range(len(fcc_data[i]["possible-values"]))):
        temp_ver[k] = fcc_data[i]["value_meaning"][k]["en"]
        temp_index[k] = n
    dt[i] = [fcc_data[i]["question_en"], dt_str_end[i], len(fcc_data[i]["possible-values"]),
           temp_ver, temp_index]
  else:
    dt[i] = [fcc_data[i]["question_en"], dt_str_end[i], len(fcc_data[i]["possible-values"]),
           None, None]
  count += dt[i][2] + 1
  dt_list.append(dt[i])
```

```python
import json


with open('drive/MyDrive/disease/info.json', 'w') as file:
    json.dump(dt, file)
```
